```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
```

```python
df  = pd.read_excel("/content/netflix_titles_agoravai123.xlsx")
```

```python
## aplicando drop nas colunas que não está sendo usada
df.drop('lançamento', 1)
```

---
# **`Gráfico Tipo x Show Id`**

```python
dftipo = df.groupby('type')[['show_id']].count()
dftipo
```

```python
dftipo.plot(kind='bar',color='g')
```

---
# **`Gráfico Release Year  x Type`** 

```python
dfReano = df.groupby('release_year')[['type']].count().tail(30)
dfReano
```

```python
dfReano.plot(kind='bar',color='g')
```

---
# **`Gráfico Duration x Type (Movie/TV Show)`** 

```python

dfduracao = df.groupby('duration')[['type']].count().tail(20)
dfduracao
```

```python
dfduracao.plot(kind='bar',color='g')
```

---
# **`Gráfico Listed In (categoria) x Type (Movie/TV Show)`** 

```python
dflista = df.groupby('listed_in')[['type']].count()
dflista
```

```python
dflista.plot(kind='barh',color='g')
```
